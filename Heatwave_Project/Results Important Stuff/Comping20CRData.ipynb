```python
import xarray as xr

# Assuming you have the individual datasets TMAX, TMIN, PRMSL, APCP, SHUM, UWND1000, VWND1000, UWND850, VWND850, UWND500, VWND500, UWND200, VWND200

# Step 1: Remove conflicting attributes before the merge
variables = [TMAX, PRMSL, APCP, SHUM, UWND1000, VWND1000, UWND850, VWND850, UWND500, VWND500, UWND200, VWND200]
for var in variables:
    var.attrs['_FillValue']= np.nan
    var.attrs['missing_value']= np.nan

# Step 2: Perform the merge operation
CR20_INDO_PAC_REG = xr.merge(variables)

# Step 3: Set consistent _FillValue and missing_value attributes for all variables
fill_value = -9.969209968386869e+36
for var in CR20_INDO_PAC_REG.variables:
    if '_FillValue' not in CR20_INDO_PAC_REG[var].attrs:
        CR20_INDO_PAC_REG[var].attrs['_FillValue'] = fill_value
    if 'missing_value' not in CR20_INDO_PAC_REG[var].attrs:
        CR20_INDO_PAC_REG[var].attrs['missing_value'] = fill_value

# Continue with your remaining code...
CR20_INDO_PAC_REG.to_netcdf(r"E:\LIBRARY\UNIVERSITY\Masters Research\20CR\20_CR_Datasets\20CR_DATA.nc")

```
