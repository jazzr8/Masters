```python
#Define the Sept as a function
def linear_findings_MM_Subs(Month_Data):
    #Now develop the looping fields
    #The Month Used
    Month_U= Month_Data


    #Drop nans
    #Max is 9 to 9
    #Reset Index so the index is looped not the date

    i = ['0','3','6','9','12','15','18','21']
    j = ['PRO Max', 'PRO Min']

    #Most likely the Min will be between 12am and 9am
    #Most likely the Max wil be between 9am to 6pm


    a_storage= []
    b_storage = []
    corr = []
    fig, axs = plt.subplots(nrows = len(i), ncols = len(j), sharey = True, sharex = True,figsize = (21,20))

    #Let's begin by extracting the Hour First
    for I in range(len(i)):
        hour =  i[I]
        #Extract Sub Daily
        Sub_Daily = Month_U[hour].dropna()
    
        for J in range(len(j)):
            #Extract Extreme Temperatures
            MM = j[J]
            Daily_Ext = Month_U[MM]
        
            #Now combine them together so its the same length
        
            DS =  pd.merge(left = Sub_Daily, right  =Daily_Ext, left_index=True,right_index=True)
        
            #Plot The subplot and see if it works
            axs[I,J].scatter(DS[hour], DS[MM])  
            #axs[I,J].set_xlabel(hour)
            axs[I,J].set_ylabel(MM)
        
            #Now develop the linear regression equation
            A, B = np.polyfit(DS[hour], DS[MM], 1)
            a_storage.append(A)
            b_storage.append(B)
            #Find the correlation                  
            corr, _ = pearsonr(DS[hour], DS[MM])    
            #Plot the linear regression line on the axs
            axs[I,J].plot(DS[hour], A*DS[hour]+B,color ='black',label = '{} = {}*(Time:{})+({}), r\N{SUPERSCRIPT TWO}: {}'.format(MM,round(A, 2),hour,round(B, 2),round(np.power(corr,2), 3)))   
            #Plot the Legend
            axs[I,J].legend()

    Coef_A = pd.Series(a_storage,name = 'A')
    Coef_B = pd.Series(b_storage,name = 'B')
    Coef_Corr = pd.Series(a_storage,name = 'Corr')
    Imp_Values = pd.concat([Coef_A,Coef_B,Coef_Corr],axis=1)
    Max_Values = pd.concat([Imp_Values.loc[0],Imp_Values.loc[2],Imp_Values.loc[4],Imp_Values.loc[6],Imp_Values.loc[8],Imp_Values.loc[10],Imp_Values.loc[12],Imp_Values.loc[14]],axis=1)
    Min_Values = pd.concat([Imp_Values.loc[1],Imp_Values.loc[3],Imp_Values.loc[5],Imp_Values.loc[7],Imp_Values.loc[9],Imp_Values.loc[11],Imp_Values.loc[13],Imp_Values.loc[15]],axis=1)
    return(fig,Max_Values,Min_Values)
```
